```python
# Step 1: Install and Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Upload Your Dataset
from google.colab import files
uploaded = files.upload()

# Step 3: Load Data
df = pd.read_csv(next(iter(uploaded)))
df.head()
```

```python
# Step 4: Filter for One Partner (e.g., Partner A)
partner = "Partner A"
partner_df = df[df["Partner"] == partner].copy()
partner_df.reset_index(drop=True, inplace=True)

# Add Month Index for Regression
partner_df["MonthIndex"] = np.arange(len(partner_df))

# Step 5: Linear Regression for Revenue
X = partner_df[["MonthIndex"]]
y = partner_df["Revenue"]
model = LinearRegression().fit(X, y)

# Step 6: Forecast for July 2025 (next month)
next_month_index = len(partner_df)
forecast = model.predict([[next_month_index]])

print(f"🔮 Forecasted Revenue for July 2025 ({partner}): ₹{forecast[0]:,.2f}")
```

```python
# Step 7: Plot Trend
plt.figure(figsize=(10, 5))
plt.plot(partner_df["MonthIndex"], y, label="Actual Revenue")
plt.plot(partner_df["MonthIndex"], model.predict(X), label="Trend Line", linestyle="--")
plt.scatter(next_month_index, forecast, color='red', label="July 2025 Forecast")
plt.title(f"{partner} - Revenue Trend & Forecast")
plt.xlabel("Month Index")
plt.ylabel("Revenue")
plt.legend()
plt.grid(True)
plt.show()
```
